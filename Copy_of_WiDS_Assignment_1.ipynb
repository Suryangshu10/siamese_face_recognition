## ðŸ§© Task 2: Image Preprocessing (OpenCV / Matplotlib)

**Goal:** Prepare images in the format required by the Siamese Network (105Ã—105 grayscale, normalized).

```python
import cv2
import matplotlib.pyplot as plt

# TODO: Read an image and convert to Grayscale
img_gray = cv2.cvtColor(cv2.imread("image.jpg"), cv2.COLOR_BGR2GRAY)

# TODO: Resize to 105x105 as required by the Siamese Paper
img_resized = cv2.resize(img_gray, (105, 105))

# TODO: Normalize pixel values to be between 0 and 1
img_final = img_resized / 255.0

plt.imshow(img_final, cmap='gray')
plt.title("Ready for the Network!")
plt.show()
```

## ðŸ§© Task 3: Building the Twin Network (TensorFlow)

**Goal:** Build the feature extractor used in Siamese Networks.

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def create_tower():
    model = models.Sequential([
        # TODO: Add a Convolution layer (64 filters, 10x10 kernel)
        layers.Conv2D(64, (10, 10), activation='relu', input_shape=(105, 105, 1)),

        # TODO: Add Max Pooling (2x2)
        layers.MaxPooling2D((2, 2)),

        layers.Flatten(),

        # TODO: Add the final Dense layer (4096 units)
        layers.Dense(4096, activation='sigmoid')
    ])
    return model

my_tower = create_tower()
my_tower.summary()
```
